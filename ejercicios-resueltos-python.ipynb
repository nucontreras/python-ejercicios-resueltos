El número de alumno tiene un dígito verificador. Para todo número de alumno, el último dígito corresponde al verificador y puede calcularse a partir de los primeros n-1 dígitos.

El algoritmo para calcular el dígito verificador es el siguiente:

1. Multiplicar cada dígito por 2,3,...,8,2,3... de atrás hacia adelante, por ejemplo si los digitos ingresados son 1863921, la multiplicacion sería:


        1 * 2 = 2
        2 * 3 = 6
        9 * 4 = 36
        3 * 5 = 15
        6 * 6 = 36
        8 * 7 = 56
        1 * 8 = 8

2. Sumar las multiplicaciones parciales

    2 + 6 + 36 + 15 + 36 + 56 + 8 = 159
    
3. Calcular el resto de la división por 11

    159 % 11 = 5

4. El dígito verificador es 11 menos el resultado anterior. Si la resta es 10, se cambia por una 'j' y, si es 11, se reemplaza por 0.

    11 - 5 = 6

**Tarea**

Escribir una función que reciba los primeros n-1 dígitos de un número de alumno y retorne el dígito verificador.

**Input Format**

Un numero entero de 7 digitos

**Constraints**

None

**Output Format**

" El digito verificador es : X "

donde X es el digito computado por tu programa.

---

Link al problema: 

Creditos....

`````
""" Digito Verificador """

numero = input()

def Verificador(numero):
    
    reves = numero[::-1]
    suma = 0
    factor = 2 
    
    for digito in reves:
        producto =  int(digito) * factor
        suma += producto
        factor += 1
        
    resto = suma % 11
    resultado = 11 - resto 
    
    if resultado == 10:
        verificador = "j"
        
    elif resultado == 11:
        verificador = 0 
        
    else:
        verificador = resultado
        
    return verificador 

print("El digito verificador es:", Verificador(numero))
`````

Recibe como parámetro un número entero ``` x = x_0, x_1, ..., x_n```. Debe:

**Objetivos**

**El Problema**

Eres una hormiga que busca planificar una excursión a la parte más alta del árbol en el que vives 🐜 🎒 🌳. Lamentablemente eres muy pequeña y desde el suelo no puedes distinguir cómo llegar allá.

Representamos una rama como una lista de elementos. Cada elemento en una rama puede ser:

es un árbol de altura máxima 5, con una rama representada por `[2,3,4,5]`.

Dado un árbol, el camino para llegar al punto más alto se representa por las elecciones en el árbol. Por ejemplo, para el árbol anterior, para llegar a la altura máxima 5:

Dado un árbol, el camino para llegar al punto más alto se representa por las elecciones en el árbol. Por ejemplo, para el árbol anterior, para llegar a la altura máxima 5:

Representamos este camino como `[2,3]`.

**Tu Tarea**

Recibirás como entrada una línea que representa un árbol. Debes implementar un algoritmo de *Backtracking* que encuentre el camino hacia el punto más alto en el árbol. Finalmente debes imprimir una línea con el camino hacia el punto más alto del árbol.

**HINT:** Recuerda que `type(var)` retorna el tipo de dato de una variable, por lo que es posible compararlo con `int` o `list` (sin comillas) para ver si efectivamente corresponde a ese tipo de dato.

Link al problema: [https://www.hackerrank.com/contests/iic1103-2017-lab11/challenges/la-rama-mas-alta/problem](https://www.hackerrank.com/contests/iic1103-2017-lab11/challenges/la-rama-mas-alta/problem)

`````
""" La rama más alta """

import ast


class Arbol:
    def __init__(self, arbol):
        self.arbol = arbol
        self.hoja_mas_alta = 0
        self.solucion = []

    def backtracking(self, arbol, camino_actual=[]):
        i = 0
        while i < len(arbol):
            if type(arbol[i]) == list:
                camino = camino_actual.copy()
                camino.append(i)
                self.backtracking(arbol[i], camino)
            elif (i + 1) == len(arbol) and type(arbol[i]) == int:
                if arbol[i] > self.hoja_mas_alta:
                    camino = camino_actual.copy()
                    camino.append(i)
                    self.hoja_mas_alta = arbol[i]
                    self.solucion = camino
            i += 1


# Función que encuentra el camino hacia el punto más alto de un árbol
def camino(arbol):
    arbol_completo = Arbol(arbol)
    arbol_completo.backtracking(arbol_completo.arbol)
    return arbol_completo.solucion


#El iguiente código recibe la entrada y lo transforma al formato descrito. Se sugiere no modificarlo.
arbol = ast.literal_eval(input())
print(camino(arbol))
`````

`````
""" Búsqueda del tesoro """


class Celda:
    def __init__(self, celda, posicion):
        self.tipo = celda
        self.posicion = posicion
        self.revisada = False

    def __str__(self):
        return self.tipo

class Laberinto:
    def __init__(self, laberinto):
        self.matriz = laberinto
        self.solucion = False
        self.camino = []
        self.ancho = len(laberinto[0])
        self.alto = len(laberinto)

    def encontrar_tesoro(self, pos=(0,0), ruta_actual=[]):
        # Backtracking.
        if self.solucion:
            return
        elif pos[0] < 0 or pos[1] < 0 or pos[0] > (self.alto - 1) or pos[1] > (self.ancho - 1):
            return
        celda = self.matriz[pos[0]][pos[1]]
        if celda.revisada:
            return
        elif celda.tipo == "0":
            return
        if not celda.revisada:
            ruta = ruta_actual.copy()
            if celda.tipo == "T":
                print("Encuentra solución")
                self.solucion = True
                ruta.append(celda.posicion)
                self.camino = ruta
            else:
                # Recursión
                celda.revisada = True
                ruta.append(celda.posicion)
                self.encontrar_tesoro((pos[0], pos[1] + 1), ruta)
                self.encontrar_tesoro((pos[0] + 1, pos[1]), ruta)
                self.encontrar_tesoro((pos[0] - 1, pos[1]), ruta)
                self.encontrar_tesoro((pos[0], pos[1] - 1), ruta)  # hay que arreglar este loop

    def __str__(self):
        string = ""
        for tupla in self.camino:
            string += str(tupla[0]) + "," + str(tupla[1]) + " > "
        return string.strip(" > ")


def adaptar_input(input):
    input_ = input
    matriz_input = input_.split("/")
    matriz = []
    for elem in matriz_input:
        matriz.append(elem.split(" "))
    return matriz

def cargar_laberinto(matriz):
    matriz_ = []
    for row in range(len(matriz)):
        fila = []
        for col in range(len(matriz[0])):
            celda = Celda(matriz[row][col], (row, col))
            fila.append(celda)
        matriz_.append(fila)
    laberinto = Laberinto(matriz_)
    return laberinto

matriz_string = input()
matriz = adaptar_input(matriz_string)
laberinto = cargar_laberinto(matriz)
laberinto.encontrar_tesoro()
print(laberinto)
`````

**    Aplicar conceptos de backtracking a un problema.

**El Problema**

Desde que comenzó el universo, la suma de subconjuntos ha sido un problema clásico en la computación. Dado un conjunto de números $C$, el problema consiste en encontrar todos los subconjuntos $C_i \subset C$ cuya suma de sus elementos es un número arbitrario $n$. Esto es: 

$$\left\lbrace C_i: \left( \sum_{x \in C_i} x \right) = n \right\rbrace$$

Solucionar este problema puede ser muy difícil, ya que existe una cantidad exponencial de posibles subconjuntos de C. Por esto, debes utilizar *backtracking* para solucionarlo.

**Input**

Recibirás como entrada una línea con un número entero **n** que representa la suma de subconjuntos que representa el largo exacto del subconjunto y luego una línea con una lista de números enteros separados por un espacio. Por ejemplo:
````
4
1 3 5 6 7 8 2
````

**Tu Tarea**

Debes implementar un algoritmo que encuentre un subconjunto de números que suman **n**, utilizando backtracking. Debes imprimir el subconjunto **encontrado** ordenado de menor a mayor.

**Hint**: En este tipo de ejercicios en los que hay que probar muchas combinaciones distintas, conviene deshacerse de las combinaciones que ya sabemos que no van a funcionar. Por ejemplo, si la lista es `0,10,20,5,7,1,2` y `R=11`, cualquier combinación que tenga al 20 sabes que no va a funcionar, así que puedes descartarlas sin probarlas.

**Hint 2**: Muchas veces resulta útil evaluar un objeto cualquiera en una sentencia if. En python, if (x) interpreta x como False cuando es un string vacio, lista vacía, None, o simplemente False, y True en cualquier otro caso. De esta forma puedes programar tus casos base con más que simplemente booleans.

**Input Format**

Una línea con un número entero **n** y luego una línea con una lista de números enteros separados por un espacio.

**Constraints**

La lista sólo contiene números enteros positivos y hay al menos una combinación de números correcta.

**Output Format**

La lista ordenada con el subconjunto pedido

Link al problema: [https://www.hackerrank.com/contests/iic1103-2017-lab11/challenges/suma-subset/problem](https://www.hackerrank.com/contests/iic1103-2017-lab11/challenges/suma-subset/problem)


`````
""" Suma de subconjuntos """


class Subconjunto:
    def __init__(self, lista, total):
        self.lista = lista
        self.total = total
        self.solucion = False
        self.resultado = []

    def subset(self, lista, opcion_actual=[]):
        # Backtracking
        if self.solucion:
            return
        suma_actual = sum(opcion_actual)
        lista_ = lista.copy()
        for numero in lista:
            lista_.pop(lista_.index(numero))
            if (suma_actual + numero) == self.total:
                opcion = opcion_actual.copy()
                opcion.append(numero)
                opcion.sort()
                self.resultado = opcion
                self.solucion = True
            elif (suma_actual + numero) < self.total:
                opcion = opcion_actual.copy()
                opcion.append(numero)
                self.subset(lista_, opcion)


def numeros_deseados(lista, total):
    nueva_lista = []
    for numero in lista:
        if numero <= total:
            nueva_lista.append(numero)
    return nueva_lista



n = int(input())  # Suma de subconjuntos
lista = [int(x) for x in input().split()]  # Lista de números

nueva_lista = numeros_deseados(lista, n)
subconjunto = Subconjunto(nueva_lista, n)
subconjunto.subset(subconjunto.lista)

print(subconjunto.resultado)
`````
